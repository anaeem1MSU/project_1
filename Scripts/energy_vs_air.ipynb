```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy.stats as stats
```

```python
# import general clean Dataframe

general_data = pd.read_csv("../Outputs/cleaned_data.csv")

#Import energy use DataFrames

United_energy_df = pd.read_csv("../Outputs/United.csv")
Mexico_energy_df = pd.read_csv("../Outputs/Mexico.csv")
India_energy_df = pd.read_csv("../Outputs/India.csv")
Japan_energy_df = pd.read_csv("../Outputs/Japan.csv")
France_energy_df = pd.read_csv("../Outputs/France.csv")
Germany_energy_df = pd.read_csv("../Outputs/Germany.csv")
Spain_energy_df = pd.read_csv("../Outputs/Spain.csv")
Brazil_energy_df = pd.read_csv("../Outputs/Brazil.csv")
Australia_energy_df = pd.read_csv("../Outputs/australia.csv")
Italy_energy_df = pd.read_csv("../Outputs/italy.csv")
```

```python
# Compute linear regression and a scatterplot for every country
linear_regression = linregress(general_data['Non-Renewables Value'], general_data['Particulate Matter'])
linear_equation = linear_regression[0] * general_data['Non-Renewables Value'] + linear_regression[1]
line_eq = f'y = {str(round(linear_regression[0],2))} x + {str(round(linear_regression[1],2))}'
print(linear_regression)

#print(f'The r-value is: {round(linear_regression[2],3)}')
general_data.plot.scatter(
    x = 'Non-Renewables Value',
    y = 'Particulate Matter',)

plt.plot(general_data['Non-Renewables Value'],linear_equation,color='black')

plt.title('Non-Renewables vs. Particulate Matter for All Countries')
plt.xlabel('Non-Renewables Value (GW/h)')
plt.ylabel('Particulate Matter (µg/m³)')
plt.xlim(-1000,75000)
plt.ylim(0,30)
plt.grid()
plt.show()
```
